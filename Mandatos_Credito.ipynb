```python
def identifica_mandato(nome,tipo):
    nome = nome.lower()
    tipo = tipo.lower()
    if 'infra' in nome and 'exclusivo' in tipo:
        tipo = 'Infra_Exclusivo'
    elif 'estrutu' in nome and 'exclusivo' in tipo:
        tipo = 'Estruturado_Exclusivo'    
    elif 'corporat' in nome and 'exclusivo' in tipo:
        tipo = 'Corporativo_Exclusivo'         
    else:
        tipo = 'não_aplicado'
    return tipo     
```
